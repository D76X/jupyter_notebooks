```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import norm
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style='darkgrid')
```

```python
# loc: mean=0, scale: std dev=1
normally_distributed_data = np.random.normal(loc = 0, scale = 1, size =100000)
```

```python
normally_distributed_data[:10]
```

```python
sample_mean = np.mean(normally_distributed_data)
```

```python
sample_mean
```

```python
# The mean and the median of an ND are always very close as there are not many outliers
sample_med = np.median(normally_distributed_data)
sample_med
```

```python
sample_std = np.std(normally_distributed_data)
sample_std
```

```python
plt.figure(figsize=(12,8))
sns.histplot(x=normally_distributed_data)
plt.title('Normal Distribution', fontsize='15')
plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='r', linestyle='-')
plt.xlabel('Values', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(x=normally_distributed_data)
plt.title('Normal Distribution', fontsize='15')
plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='r', linestyle='-')
plt.xlabel('Values', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

```python
# loc: mean=0, scale: std dev=3
normally_distributed_data_scale_3 = np.random.normal(loc = 0, scale = 3, size = 100000)
```

```python
plt.figure(figsize=(12,8))

# ND # loc: mean=0, scale: std dev=1
sns.kdeplot(x=normally_distributed_data)

# ND # loc: mean=0, scale: std dev=3
sns.kdeplot(x=normally_distributed_data_scale_3)

plt.title('Normal Distributions', fontsize='15')

plt.axvline(x=sample_mean, color='yellow', linestyle='-')
plt.axvline(x=sample_med, color='r', linestyle='-')

plt.xlabel('Values', fontsize='15')
plt.ylabel('Probability', fontsize='15')

plt.show()
```
